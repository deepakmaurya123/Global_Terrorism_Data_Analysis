```python
df = df[df["country"] < 1000]
df = df[df["nationality"] < 600]
df = df[df["nkill"] < 518]
#df = df[df['gname'] != 'Unknown']
df["nkill"] = df["nkill"].astype(int) 
```

```python
df.info()
```

# Eploratory Data Analysis

```python
no_kill = df.groupby(["year"])["nkill"].sum()
no_kill
```

```python
plt.figure(figsize=(13,5))

p1 = sns.barplot(x=no_kill.index,y=no_kill, palette='hsv')
p1.set_xticklabels(no_kill.index, rotation = 90)
plt.title("Number of casualities every year (1970 - 2014)" ,fontsize = 20)
plt.xlabel('YEAR')
plt.ylabel('CASUALITY')
plt.savefig("image_1.jpg")

plt.show()
```

```python
no_of_crime = df["year"].value_counts()
no_of_crime.sort_index(inplace = True)
no_of_crime.fillna(0)
```

```python
plt.figure(figsize=(13,5))
sns.lineplot(x = no_of_crime.index, y= no_of_crime, color='red' ,marker='o')
plt.title("Number of crime yearwise" ,fontsize = 20)
plt.ylabel('Crime Number')
plt.xlabel('Year')
plt.savefig("image_2.jpg")
plt.show()
```

```python
top_10_country = df["country_txt"].value_counts()[:10]
top_10_country
```

```python
plt.figure(figsize=(13,5))
#c = ['red', 'yellow', 'green', 'blue', 'orange', 'cyan' ,'magenta']
c = ["#e82727","#ff8649","#ffc549","#fced20","#8fe222","#22e28e","#22e2db","#1f85c4","#361599","#d11fce"]
plt.bar(top_10_country.index, top_10_country ,color=c, width=0.5)
plt.title("Top 10 country in crime" ,fontsize = 20)
plt.xlabel('No of Crime')
plt.ylabel('Country')
plt.savefig("image_3.jpg")
plt.show()
```

```python
df_region = df["region_txt"].value_counts()[:8]
df_region
```

```python
plt.figure(figsize=(5,5))
plt.title("Crime percent with region in this world" ,fontsize = 20)
plt.pie(df_region, autopct = '%.2f', labels = df_region.index)
plt.savefig("image_4.jpg" ,bbox_inches = "tight")
plt.show()
```

```python
target_people = df[df["region_txt"] == 'South Asia']['targtype_txt'].value_counts()
target_people
```

```python
attack_type = df[df["region_txt"] == 'South Asia']['attacktype1_txt'].value_counts()[:7]
attack_type
```

```python
plt.figure(figsize=(13,6))

plt.title("Different Target type in South Asia", fontsize=20)

sns.barplot(y = target_people.index,x=target_people, palette='gist_rainbow')
plt.ylabel("targeted people")
plt.savefig("image_5.jpg" ,bbox_inches = "tight")
```

```python
plt.figure(figsize=(10,3))
plt.title("title",fontsize=20)
plt.title("Different Attack type in South Asia", fontsize=20)
ax = sns.barplot(x = attack_type.index, y=attack_type, width = 0.5, palette='bright')
plt.xlabel("Attacked Type")
plt.ylabel("No of attack")
ax.set_xticklabels(attack_type.index, rotation = -80)
plt.savefig("image_6.jpg" ,bbox_inches = "tight")
plt.show()
```

```python
top_10_gp = df[df["gname"] != "Unknown"]["gname"].value_counts()[:10]
top_10_gp
```

```python
plt.figure(figsize=(10,5))
sns.barplot(y = top_10_gp.index, x = top_10_gp, palette='gist_rainbow')
plt.title("Top 10 terrorist group in this world", fontsize=20)
plt.xlabel("No of Crime")
plt.ylabel("Terrorist Group")
plt.savefig("image_7.jpg" ,bbox_inches = "tight")
```

```python
value = top_10_gp.index
s = pd.DataFrame()
for i in value:
    data = df[df["gname"] == i].groupby("year")["gname"].count()
    s[i] = data
s.reset_index(inplace = True)
s
```

```python
c = ["#e82727","#ff8649","#ffc549","#fced20","#8fe222","#22e28e","#22e2db","#1f85c4","#361599","#d11fce"]
plt.figure(figsize=(13,9))
plt.xlabel("Year")
plt.ylabel("No of crime")
plt.title("No of crime each year by top 10 terrorist group", fontsize=20)
for i in range(10):
    plt.plot(s["year"],s[value[i]] ,color=c[i] ,label=value[i])
    plt.legend()
plt.savefig("image_8.jpg" ,bbox_inches = "tight")
```
